```python
URL = 'https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD'
```

```python
from urllib.request import urlretrieve
urlretrieve(URL,'Fremont.csv')
```

```python
#Read and output first few lines of file
import pandas as pd
data = pd.read_csv('Fremont.csv')
data.head()
```

```python
#Read and Tidy-up, then output first few lines of file
import pandas as pd
data = pd.read_csv('Fremont.csv',index_col='Date',parse_dates=True)
data.head()
```

```python
#Visualise using matplotlib
%matplotlib inline
data.plot();
```

```python
#Visualise by summing up by Week, using matplotlib
%matplotlib inline
data.resample('W').sum().plot();
```

```python
#Visualise by changing colour palette, changing labels, summing up by Week, using matplotlib
import matplotlib.pyplot as plt
plt.style.use('seaborn')
data.columns = ['West', 'East']
data.resample('W').sum().plot()
```

```python
#Represent each point as sum of previous 365 days, i.e. rolling year
data.resample('D').sum().rolling(365).sum().plot()
```

```python
#As above, but zero the axes to visualise better
ax = data.resample('D').sum().rolling(365).sum().plot()
ax.set_ylim(0, None)
```

```python
#As above, but add new combined line to show overall numbers
data['Total']=data['West']+data['East']
ax = data.resample('D').sum().rolling(365).sum().plot()
ax.set_ylim(0, None)
```

```python
#Plot data by time of day across the dataset
data.groupby(data.index.time).mean().plot()
```

```python
pivoted = data.pivot_table(['Total'], index=data.index.time, columns=data.index.date)
```
